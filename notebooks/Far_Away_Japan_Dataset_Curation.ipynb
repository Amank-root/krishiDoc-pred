```python
import os
import shutil
import re
from pathlib import Path
from tqdm import tqdm

# ============================================================
# SOURCE DATASETS
# ============================================================

DATASETS = [
    "/content/datasets/crops-disease-dataset",
    "/content/datasets/Crop___Disease",
    "/content/datasets/PlantVillage",
]

FINAL_ROOT = "/content/datasets/final_dataset"

# ============================================================
# CLASS MAPPING
# Final format = Crop_Disease
# ============================================================

CLASS_MAP = {

    # ---------------- CORN ----------------
    "Corn___Healthy": "Corn_Healthy",
    "Corn___Common_Rust": "Corn_Common_Rust",
    "Corn___Gray_Leaf_Spot": "Corn_Gray_Leaf_Spot",
    "Corn___Northern_Leaf_Blight": "Corn_Northern_Leaf_Blight",

    # ---------------- POTATO ----------------
    "Potato___Healthy": "Potato_Healthy",
    "Potato___healthy": "Potato_Healthy",

    "Potato___Early_Blight": "Potato_Early_Blight",
    "Potato___Early_blight": "Potato_Early_Blight",

    "Potato___Late_Blight": "Potato_Late_Blight",
    "Potato___Late_blight": "Potato_Late_Blight",

    # ---------------- RICE ----------------
    "Rice_Healthy": "Rice_Healthy",
    "Rice___Healthy": "Rice_Healthy",

    "Rice_BrownSpot": "Rice_Brown_Spot",
    "Rice___Brown_Spot": "Rice_Brown_Spot",

    "Rice_Hispa": "Rice_Hispa",

    "Rice_LeafBlast": "Rice_Leaf_Blast",
    "Rice___Leaf_Blast": "Rice_Leaf_Blast",

    "Rice___Neck_Blast": "Rice_Neck_Blast",

    # ---------------- WHEAT ----------------
    "Wheat_Healthy": "Wheat_Healthy",
    "Wheat___Healthy": "Wheat_Healthy",

    "Wheat_BrownRust": "Wheat_Brown_Rust",
    "Wheat___Brown_Rust": "Wheat_Brown_Rust",

    "Wheat_YellowRust": "Wheat_Yellow_Rust",
    "Wheat___Yellow_Rust": "Wheat_Yellow_Rust",

    "Wheat_BlackRust": "Wheat_Black_Rust",
    "Wheat_Blast": "Wheat_Blast",
    "Wheat_Aphid": "Wheat_Aphid",
    "Wheat_CommonRootRot": "Wheat_Common_Root_Rot",
    "Wheat_FusariumHeadBlight": "Wheat_Fusarium_Head_Blight",
    "Wheat_LeafBlight": "Wheat_Leaf_Blight",
    "Wheat_Mildew": "Wheat_Mildew",
    "Wheat_Mite": "Wheat_Mite",
    "Wheat_Septoria": "Wheat_Septoria",
    "Wheat_Smut": "Wheat_Smut",
    "Wheat_Stemfly": "Wheat_Stemfly",
    "Wheat_Tanspot": "Wheat_Tan_Spot",

    # ---------------- SUGARCANE ----------------
    "Healthy": "Sugarcane_Healthy",
    "Bacterial Blight": "Sugarcane_Bacterial_Blight",
    "Red Rot": "Sugarcane_Red_Rot",

    # ---------------- PEPPER ----------------
    "Pepper__bell___healthy": "Pepper_Healthy",
    "Pepper__bell___Bacterial_spot": "Pepper_Bacterial_Spot",

    # ---------------- TOMATO ----------------
    "Tomato_healthy": "Tomato_Healthy",
    "Tomato_Bacterial_spot": "Tomato_Bacterial_Spot",
    "Tomato_Early_blight": "Tomato_Early_Blight",
    "Tomato_Late_blight": "Tomato_Late_Blight",
    "Tomato_Leaf_Mold": "Tomato_Leaf_Mold",
    "Tomato_Septoria_leaf_spot": "Tomato_Septoria_Leaf_Spot",
    "Tomato_Spider_mites_Two_spotted_spider_mite":
        "Tomato_Spider_Mites",
    "Tomato__Target_Spot": "Tomato_Target_Spot",
    "Tomato__Tomato_mosaic_virus": "Tomato_Mosaic_Virus",
    "Tomato__Tomato_YellowLeaf__Curl_Virus":
        "Tomato_Yellow_Leaf_Curl_Virus",
}

# ============================================================
# CREATE OUTPUT DIRECTORY
# ============================================================

os.makedirs(FINAL_ROOT, exist_ok=True)

copied = 0
skipped = []

for dataset in DATASETS:

    dataset = Path(dataset)

    print(f"\nProcessing: {dataset}")

    for root, dirs, files in os.walk(dataset):

        folder_name = Path(root).name

        if folder_name not in CLASS_MAP:
            continue

        class_name = CLASS_MAP[folder_name]

        target_dir = Path(FINAL_ROOT) / class_name
        target_dir.mkdir(parents=True, exist_ok=True)

        images = [
            f for f in files
            if f.lower().endswith(
                (".jpg", ".jpeg", ".png", ".bmp", ".webp")
            )
        ]

        for img in tqdm(images, leave=False):

            src = Path(root) / img

            # ====================================================
            # CLEAN FILENAME
            # ====================================================

            stem = Path(img).stem.strip()           # remove leading/trailing spaces
            stem = re.sub(r"\s+", "_", stem)        # replace multiple spaces with _
            suffix = Path(img).suffix.lower().strip()

            clean_name = f"{stem}{suffix}"

            # unique filename
            filename = f"{src.parent.name}_{clean_name}"

            dst = target_dir / filename

            try:
                shutil.copy2(src, dst)
                copied += 1

            except Exception as e:
                skipped.append((str(src), str(e)))

print("\n====================================")
print("DONE")
print("====================================")
print("Images copied :", copied)
print("Skipped       :", len(skipped))
print("Classes       :", len(os.listdir(FINAL_ROOT)))

if skipped:
    print("\nFirst 10 skipped files:")
    for item in skipped[:10]:
        print(item)
```

```python
# Exact duplicate removal (MD5) ✔️
import hashlib
from pathlib import Path
from tqdm import tqdm

DATASET = Path("/content/datasets/final_dataset")

seen_hashes = {}
duplicates = []

for img_path in tqdm(list(DATASET.rglob("*"))):
    if img_path.suffix.lower() not in {".jpg", ".jpeg", ".png", ".bmp", ".webp"}:
        continue

    with open(img_path, "rb") as f:
        file_hash = hashlib.md5(f.read()).hexdigest()

    if file_hash in seen_hashes:
        duplicates.append(img_path)
    else:
        seen_hashes[file_hash] = img_path

print(f"Found {len(duplicates)} exact duplicates")

# Delete duplicates
for dup in duplicates:
    dup.unlink()

print("Done")
```

```python
# Perceptual duplicate detection (phash, threshold=3)
from pathlib import Path
from PIL import Image
import imagehash
from tqdm import tqdm

DATASET = Path("/content/datasets/final_dataset")

hashes = []
duplicates = []

THRESHOLD = 3

images = [
    p for p in DATASET.rglob("*")
    if p.suffix.lower() in [".jpg", ".jpeg", ".png", ".bmp", ".webp"]
]

for img_path in tqdm(images):

    try:
        h = imagehash.phash(
            Image.open(img_path).convert("RGB")
        )

        duplicate_found = False

        for existing_hash, existing_path in hashes:

            if h - existing_hash <= THRESHOLD:
                duplicates.append(img_path)
                duplicate_found = True
                break

        if not duplicate_found:
            hashes.append((h, img_path))

    except Exception:
        pass

print(f"Near duplicates found: {len(duplicates)}")

for dup in duplicates:
    dup.unlink()

print("Done")
```

```python
import cv2
import random
from pathlib import Path
from tqdm import tqdm
import albumentations as A

FINAL_ROOT = Path("/content/datasets/final_dataset")
TARGET_COUNT = 500

transform = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.2),
    A.Rotate(limit=25, p=0.7),
    A.RandomBrightnessContrast(
        brightness_limit=0.2,
        contrast_limit=0.2,
        p=0.7
    ),
    A.ShiftScaleRotate(
        shift_limit=0.08,
        scale_limit=0.15,
        rotate_limit=15,
        p=0.7
    ),
    A.GaussNoise(p=0.2),
])

for class_dir in sorted(FINAL_ROOT.iterdir()):

    if not class_dir.is_dir():
        continue

    images = list(class_dir.glob("*"))

    current_count = len(images)

    if current_count >= TARGET_COUNT:
        continue

    needed = TARGET_COUNT - current_count

    print(
        f"{class_dir.name}: "
        f"{current_count} -> {TARGET_COUNT}"
    )

    for i in tqdm(range(needed), leave=False):

        src = random.choice(images)

        img = cv2.imread(str(src))

        if img is None:
            continue

        aug = transform(image=img)["image"]

        output_name = (
            f"aug_{i}_{src.stem}.jpg"
        )

        output_path = class_dir / output_name

        cv2.imwrite(str(output_path), aug)
```
